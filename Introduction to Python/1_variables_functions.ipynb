## Some Data types
There are several build-in data types in Python, which determine the type of data we want to deal with. 
For the beginning, we will consider 
- str : for textual data "hello world!", 'hi' 
- int : whole numbers $0, \pm 1, \pm 2,... $
- float : floating-point numbers (decimals), e.g. $1.2$, $-0.45$
- bool : boolean values True, False

```python
# A string (str) object can be generated by "hello" or 'hello' (equivalent)
print("hello world!")

# You can also print multiple things (not only strings)
print('hello', 'world!', 4)
```

```python
# type(a) tells you the type/class an object belongs to 
type('hello'), type(123), type(False)
```

```python
# Basic operations can be used, e.g. "+",  which works differently for different data-types
print('1+3') # no number is calculated within the string

print('1'+'3') # will glue both strings together. Still, it is a string, not a number

print(1+3) # integers are summed up and then printed

print('1+3 =', 1+3)
```

```python
# ... further operations with numbers
print(1 * 3.4, 1/2, round(1/3.4, 3)) # combination of integer and floating-point number yields floating-point number

print(1., .4) # 0 can be omitted if there is nothing in front or behind it

print(2**3) # "to the power of" is done with the ** operator

print(4**0.5) # square root
```

```python
# ... more advanced tools are provided (e.g.) within the math package
import math

print('Euler\'s number is', math.e)
print('sin(2pi) =', math.sin(2*math.pi))
```

```python
# Instances of data types can be compared -> yields a boolean value
print(1==1, 1==2, 1==1.0)

print('a' == 'a', 'a'=='A')

# ... further comparison operations
print(1>1, 1>=1, 1!=1, 1!=2)
```

## Variables

```python
# A variable is a name that refers to an object in the memory of your computer.
# The object can be an instance of a built-in data type (like int, str, list) 
# or a user-defined object (like an instance of a class).
lecture = 'ML4PDEs'
print('We participate in the lecture', lecture) # prints the attained value of the variable lecture

# A variable can be overwritten
lecture = lecture + ' is awesome' 
print(lecture)

# Variables can also be changed in-place -> but be careful in PyTorch ...
#lecture += ' is awesome' 
```

```python
# Python: Do not need to specify data type when creating variables. 
#         Type is determined during runtime -> in contrast to, e.g., Java
x = 'ML4PDEs'
x = 10  
```

```python
# Some use case of boolean values: If and Else statements
number = 1.4

if number > 1: # If the boolean value of (number > 1) is True, it will execute the command below
    print('number is larger than 1') # There must be an indent (convention 4 spaces) before every command line within the if-block

else: # Otherwise, the code below is executed
    print('number is less than or equal to 1')
```

```python
# Attention to floating-point accuracy:
print(math.sin(2*math.pi) == 0.0) # we know the sine should be 0 ... 

# ... include tolerance with math.isclose(x, y)
print(math.isclose(math.sin(2*math.pi), 0, abs_tol=1e-10)) # 1e-10 stands for 10**(-10)
```

```python
number = 1e-11

print('number is ')
if math.isclose(number, 0, abs_tol=1e-10): 
    print('almost zero')
    
elif number > 0: 
    print('positive')
    
else:
    print('negative')
```

## Methods/functions

```python
# Functions get input parameters (variables) and can do some operations with them. 
# For example, f below prints the square of the input x.
# Similar to if, else: All command lines must have the same indentation (convention: 4 spaces)
def f(x):
    print(x**2)
    print('hi') # try to shift this print-command to the right/left -> error
x1 = 12.4
f(x1)
```

```python
# Functions can of course be combined with if, else, etc. 
# -> play around with the indent in each line
def relu(x):
    if x > 0:
        return x
    else: return 0
print(relu(-1), relu(3))
```

```python
# Functions can return an output, which can be saved within some variable for further usage.
def f(x):
    return math.cos(x)

y = f(math.pi) # positional argument -> more inputs: order matters
z = f(x=math.pi) # keyword argument -> more inputs: order does not matter
print(y, z)
```

```python
# Default values of some inputs, which can sometimes be convenient
def f(x,w=1):
    return math.cos(x*w)
print( f(math.pi), f(math.pi, 1), f(math.pi, 0.5) )
```

```python
# Careful: inputs with default values must come last in function declaration
def f(w=1, x):
    return math.cos(x*w)
```

```python
# For evaluation the order of inputs does not matter if "keyword"-arguments are used:
def f(x, w=1):
    return math.cos(x*w)
f(x=math.pi, w=1), f(w=1, x=math.pi)
```

```python
# Functions can be used as inputs of other functions. For example, the method below realizes the sum of two functions
def func_sum(f, g, x):
    return(f(x)+g(x))

def f(x):
    return math.cos(x)

def h(x):
    return 2*x

func_sum(f, h, math.pi)
```

```python
# For better readability, one can also tell the user, which inputs are expected
def f(x:int, a:str):
    print(2*x, a)
f(2, 'hi')

# However, it does not prevent you from passing instances of different data types to it
f('hi', 2)
```
