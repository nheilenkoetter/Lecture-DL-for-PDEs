## Lists

```python
# A list can be used to store multiple objects in a single variable.
# Lists are created using square brackets [] and can contain items of different types.
my_list = [1, 2, 3]
print("The list is", my_list)
```

```python
# Lists can contain items of different types.
my_list = [1, "Hello", 3.14]
```

```python
# List are ordered, changeable, and allow duplicate values.
my_list = [1, True, 3, 5, 5]
print("The list is", my_list)

# We can access the items in a list by their index. The index starts at 0.
print(my_list[0]) # Output: 1
print(my_list[1]) # Output: True
print(my_list[2]) # Output: 3

# We can also use negative indexing to access the items backwards. 
# The index starts at -1 for the last item.
print(my_list[-1]) # Output: 5

# We can change the value of an item in a list by accessing it by its 
# index and assigning a new value to it.
my_list[3] = "New value"
print("Now the list is", my_list)
```

```python
# We can also slice a list to get a subset of its items. The syntax for slicing is
# list[start:stop:step]. The start index is inclusive, the stop index is exclusive.
my_list = [1, 2, 3, 4, 5]
print(my_list[1:4]) # Output: [2, 3, 4]
print(my_list[::2]) # Output: [1, 3, 5]
```

```python
# Lists are a default class in Python, we can also create them by using the 
# list() constructor.
my_list = list((1.0, 2.0, 3.0))
print(my_list)
```

```python
# Lists are mutable, we can add items to a list using the append() method.
my_list.append(False)
print("Added one value", my_list)

# We can also combine two lists using the extend() method.
my_list2 = [5.0, "Hello"]
my_list.extend(my_list2)
print("Added another list", my_list)
```

```python
# Lists can contain other lists
fruits = ["apple", "banana", "cherry"]
snacks = ["chocolate", "chips"]
shopping_list = [fruits, snacks]
print(shopping_list)

# What is the length of shopping_list?
#len(shopping_list)
```

## Loops

```python
# We can loop through the items in a list using a for loop.
for item in fruits:
    print(item)
```

```python
# We can also loop over the length of a list using a for loop and the range() function.
for i in range(len(fruits)): # what is i here?
    print(fruits[i])
```

## Plots

```python
# we have seen already how we can plot a list of y-values with plt.plot(xs, ys)
xs = [i/2 for i in range(10)]
ys = [math.cos(x) for x in xs]
zs = [math.sin(x) for x in xs]
plt.plot(xs, ys)
plt.plot(xs, zs) # multiple plots are possible within one figure
plt.show()
```

```python
# Let us illustrate a circle in the same way
zs = [math.sin(x) for x in xs]
plt.plot(ys, zs)
plt.show()
```

```python
# Alternatively, one can create scatter plots
plt.scatter(ys, zs)
plt.show()
```
