## Classes

```python
# A class is a blueprint/recipe for creating objects. 
# It defines a set of attributes and methods that the created objects will have. 
# In this example, we have a class called Car that has two attributes: 
# brand and year.
class Car:
    # A class has an initializer method called __init__ that is called when an 
    # object is created. It states the attributes of the class and initializes 
    # them with the values passed as arguments when creating an object.
    def __init__(self, brand : str, year : int) -> None:
        # self.attribute is used to refer to the attributes of the class.
        self.brand = brand
        self.year = year

    # A class can also have methods, which are functions that belong to the class.
    def info(self):
        print(f"This car is a {self.brand} from {self.year}.")
```

```python
# To create an object of the class Car, we can do the following:
car1 = Car("BMW", 2020)
car1.info()
```

```python
# We can create multiple objects of the same class with different attributes.
car2 = Car("Porsche", 1995)
car2.info()
```

```python
# Let us extend the Car class by adding:
#    - self.fuel should store the current amount of fuel in the tank
#    - self.fuel_capacity 
#    - method tank_or_drive(self, fuel_amount)
class Car():
    def __init__(self, brand : str, year : int, initial_fuel=20, fuel_capacity=50):
        self.brand = brand
        self.year = year
        self.fuel_capacity = fuel_capacity
        self.fuel = min(initial_fuel, fuel_capacity)
    
    def info(self):
        print(f"This car is a {self.brand} from {self.year}.")
        
    def check_fuel(self):
        print('Available fuel:', self.fuel)
    
    def tank_or_drive(self, fuel_amount:int):
        # driving: fuel_amount < 0 
        # tanking: fuel_amount > 0
        if fuel_amount < 0:
            self.fuel = max(self.fuel+fuel_amount, 0)
        else:
            self.fuel = min(self.fuel+fuel_amount, self.fuel_capacity)
```

```python
a = Car('Mercedes', 2005)
a.info()
a.check_fuel()
a.tank_or_drive(20)
a.check_fuel()
```

### Inheritance

```python
# Here is an analogue Bike class. The only difference is that info() prints "This bike..." instead of "This car...":
class Bike:
    def __init__(self, brand : str, year : int):
        self.brand = brand
        self.year = year

    def info(self):
        print(f"This bike is a {self.brand} from {self.year}.")
```

```python
bike = Bike("Mercedes", 2020)
bike.info()
```

```python
# Car and Bike classes have the same two attributes brand, year
#                                    method info()
# Idea: Create a superclass DrivingDevice from which Car and Bike both inherit
#       -> reduce multiple appearances of the same/similar code
#          e.g. instead of implementing the info() method for all classes Bike, Car, ...
#               implement it only a single time in the superclass DrivingDevice
class DrivingDevice():
    def __init__(self, brand : str, year : int):
        # self.attribute is used to refer to the attributes of the class.
        self.brand = brand
        self.year = year
        
    def info(self):
        print(f"This {self.__class__.__name__} is a {self.brand} from {self.year}.")
        

# The Bike class inherits all functionalities, info(), from its super class.
class Bike(DrivingDevice):
    def __init__(self, brand : str, year : int):
        # Below, we call the constructor (__init__()) of the superclass, which stores the two attributes
        super().__init__(brand, year)

# Since the Bike constructor above only calls the constructor of the super class and has the same input arguments,
# the following code is equivalent:
#class Bike(DrivingDevice):
#    pass
```

```python
bike = Bike("Mercedes", 1990)
bike.info()
```
