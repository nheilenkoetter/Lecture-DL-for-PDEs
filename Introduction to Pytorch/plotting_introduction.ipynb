```python
# For plotting we can use the matplotlib library, which is a popular plotting library 
# in Python. It has similar functionality to MATLAB's plotting functions.
import matplotlib.pyplot as plt
import torch
```

```python
x_axis = torch.linspace(0, 10, 100)
y_axis = torch.sin(x_axis)
plt.plot(x_axis.numpy(), y_axis.numpy())
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('Plot of sin(x)')
plt.show()
```

```python
# A matrix can also be interpreted as an image, where the values in the matrix represent 
# pixel intensities.
image_matrix = torch.rand(100, 100)
plt.imshow(image_matrix.numpy(), cmap='gray')
plt.show()
```
