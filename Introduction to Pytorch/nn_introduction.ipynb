```python
# In PyTorch we can find building blocks for neural networks in the torch.nn module.
import torch
import torch.nn as nn
```

```python
# A simple linear layer can be created using nn.Linear, which applies a linear 
# transformation to the input data x -> A*x + b.
linear_layer = nn.Linear(in_features=3, out_features=2)
print("Linear layer:", linear_layer)

example_input = torch.tensor([1.0, 2.0, 3.0])
output = linear_layer(example_input)
print("Output of the linear layer:", output)
```

```python
# More complex layers and architectures can be built by combining these building blocks,
# e.g. a simple feedforward neural network with one hidden layer can be defined as follows
linear_layer_1 = nn.Linear(in_features=3, out_features=5)
activation_function = nn.ReLU()
linear_layer_2 = nn.Linear(in_features=5, out_features=1)
fcn = nn.Sequential(linear_layer_1, activation_function, linear_layer_2)
print("Feedforward neural network:", fcn)

# The forward pass through the network can be performed by simply calling the network 
# on the input data.
output = fcn(example_input)
print("Output of the feedforward neural network:", output)
```

```python
# For more complex architectures, we can define our own custom nn.Module by subclassing
# nn.Module and implementing the forward method.

class CustomNN(nn.Module):
    """A simple feedforward neural network with one hidden layer."""
    def __init__(self):
        super().__init__()
        self.linear_layer_1 = nn.Linear(in_features=3, out_features=5)
        self.activation_function = nn.ReLU()
        self.linear_layer_2 = nn.Linear(in_features=5, out_features=1)

    def forward(self, x):
        x = self.linear_layer_1(x)
        x = self.activation_function(x)
        x = self.linear_layer_2(x)
        return x
    
custom_nn = CustomNN()
output = custom_nn(example_input)
print("Output of the custom neural network:", output)
```

```python
# An important aspect of neural network evaluation is the ability to apply the network
# to batches of data. This speeds up the evaluation process and allows for more efficient 
# training. 

batch_input = torch.randn(100, 3)  # A batch of 100 samples, each with 3 features
batch_output = custom_nn(batch_input)
print("Output of the custom neural network for a batch of data:", batch_output.shape)
```

### Exercise: Manual Fitting a Neural Network to given Data Points

In the previous cells the weights and biases of the linear layer were initialized randomly. In the next cell we will manually set the weights and biases to fit a given set of data points.

Our model consists of a feed forward neural network with one hidden layer. The hidden layer has 2 neurons and uses the ReLU activation function. The output layer is a linear layer that produces a single output value. Hence our model is given by
$$f(x) = w_{2,1} \cdot \text{ReLU}(w_{1,1} \cdot x + b_{1,1}) + w_{2,2} \cdot \text{ReLU}(w_{1,2} \cdot x + b_{1,2}) + b_2$$

where $w_{1,1}, w_{1,2}$ and $b_{1,1}, b_{1,2}$ are the weights and biases of the hidden layer, and $w_{2,1}, w_{2,2}$ and $b_2$ are the weights and biases of the output layer.
