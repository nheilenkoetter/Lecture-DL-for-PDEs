```python
# External Packages can be imported using the import statement. 
# We will use the PyTorch library in this course, which is a popular deep learning framework.
import torch
```

```python
# Inside torch there is a class called Tensor, which is the main data structure 
# used in PyTorch. A tensor is a multi-dimensional array that can be used to store 
# data and perform operations on it.
tensor_1 = torch.tensor([1.0, 2.0, 3.0])
print("The tensor is", tensor_1)

# Note: Tensor have to be of one data type unlike lists!
```

```python
# Tensors can be accessed and manipulated using indexing and slicing, similar to lists.
print("The first element of the tensor is", tensor_1[0])
print("The last element of the tensor is", tensor_1[-1])
print("The first two elements of the tensor are", tensor_1[:2])
```

```python
# Mathematical operations can be performed on tensors, such as addition, multiplication, etc.
tensor_2 = torch.tensor([4.0, 5.0, 6.0])
print("The sum of the two tensors is", tensor_1 + tensor_2)
print("The (elementwise) product of the two tensors is", tensor_1 * tensor_2)
```

```python
# We can create multi-dimensional tensors as well. For example, a 2D tensor (matrix)
# can be created as follows:
matrix_1 = torch.tensor([[1.0, 2.0, 0.0], [3.0, 4.0, 5.0]])
print("The matrix is", matrix_1)

# There are also construction functions to create tensors of specific shapes and values. 
# For example, we can create a tensor of zeros or random values:
zeros_tensor = torch.zeros((2, 3))
random_tensor = torch.rand((2, 2))
print("Tensor of zeros:", zeros_tensor)
print("Tensor of random values:", random_tensor)
```

```python
# Slicing can also be carried out on multi-dimensional tensors.
print("The matrix is", matrix_1)
print("Indexing with [0, 0] gives", matrix_1[0, 0])
print("Indexing with [0] gives", matrix_1[0])
print("Indexing with [:, 0] gives", matrix_1[:, 0])
print("Indexing with [1, :2] gives", matrix_1[1, :2])
```

```python
# Each tensor has a shape, which can be accessed using the .shape attribute. 
# The shape is a tuple that indicates the size of each dimension of the tensor.
print("The shape of tensor_1 is", tensor_1.shape)
print("The shape of matrix_1 is", matrix_1.shape)
```

```python
# Further torch provides a wide range of mathematical functions and operations that can be 
# applied to tensors, e.g.:
print("The mean of tensor_1 is", torch.mean(tensor_1))
print("Elementwise sine of tensor_1 is", torch.sin(tensor_1))
print("The matrix product", torch.matmul(random_tensor, random_tensor))
```

```python
# A tensor can also be moved to a GPU for faster computation if a GPU is available.
if torch.cuda.is_available():
    tensor_1 = tensor_1.to('cuda')
    print("Tensor moved to GPU:", tensor_1)

# We can check the device of a tensor using the .device attribute.
print("The device of tensor_1 is", tensor_1.device)
print("The device of matrix_1 is", matrix_1.device)
```

```python
# Tensors on different devices cannot be combined in operations. 
# For example, if tensor_1 is on the GPU and matrix_1 is on the CPU, 
# we cannot perform operations between.
result = tensor_1 + matrix_1
```

```python
# Using the GPU can speed up computations significantly.
import time

for device in ["cpu", "cuda"] if torch.cuda.is_available() else ["cpu"]:
    # Create two large tensors on the specified device
    a = torch.rand((10000, 10000), device=device)
    b = torch.rand((10000, 10000), device=device)

    # Perform a matrix multiplication and time it
    start_time = time.time()
    result = torch.matmul(a, b)
    print("Matrix multiplication on {} completed, took {:.2f} seconds".format(device.upper(), time.time() - start_time))
```
